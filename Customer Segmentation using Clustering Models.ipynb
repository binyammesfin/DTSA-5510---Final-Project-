### **3.1 Baseline Model: K-Means Clustering**

K-Means clustering is one of the most popular and widely used clustering algorithms. It partitions the dataset into K distinct, non-overlapping subsets (clusters) based on the distance between data points and the centroid of each cluster. The goal is to minimize the within-cluster variance (sum of squared distances between data points and their corresponding cluster centroids).

**Rationale for Using K-Means Clustering**
K-Means is a simple and efficient algorithm that works well with large datasets and provides easily interpretable results. It serves as a good starting point for clustering analysis because it helps to identify the general structure and distribution of the data. By setting different values of K, we can explore how the number of clusters affects the clustering results.

**Implementation**
Here, we will implement K-Means clustering on the preprocessed data and visualize the resulting clusters.

```python
# Set a random seed for reproducibility
random_seed = 42

# Select the number of clusters (K)
k = 5

# Initialize and fit the K-Means model
kmeans = KMeans(n_clusters=k, random_state=random_seed)
kmeans.fit(data_scaled)

# Add the cluster labels to the original data
data['Cluster'] = kmeans.labels_

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_scaled['Income'], y=data_scaled['MntWines'], hue=data['Cluster'], palette='viridis')
plt.title('K-Means Clustering Results')
plt.xlabel('Income (Normalized)')
plt.ylabel('MntWines (Normalized)')
plt.show()
```

### **3.2 Hierarchical Clustering**

Hierarchical clustering is an agglomerative clustering method that builds a hierarchy of clusters by progressively merging smaller clusters into larger ones. The process starts with each data point as an individual cluster and then merges the two closest clusters at each step, based on a chosen linkage criterion (e.g., single, complete, average).

**Rationale for Using Hierarchical Clustering**

Hierarchical clustering is useful for identifying nested clusters and understanding the hierarchical structure of the data. It does not require specifying the number of clusters in advance and provides a dendrogram, a tree-like diagram that visualizes the merging process and helps in determining the optimal number of clusters.

**Implementation**

Here, we will implement hierarchical clustering on the preprocessed data and visualize the resulting dendrogram.

```python
# Perform hierarchical clustering
Z = linkage(data_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

### **3.3 DBSCAN Clustering**

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm that identifies clusters as dense regions of data points separated by regions of lower density. Unlike K-Means, DBSCAN does not require specifying the number of clusters in advance and can effectively handle clusters of arbitrary shapes and sizes, as well as noise (outliers).

**Rationale for Using DBSCAN Clustering**

DBSCAN is particularly useful for data with varying density and noisy data points. It can identify core samples, border samples, and noise, making it suitable for complex datasets where other clustering methods might struggle to define clusters accurately.

**Implementation**

Here, we will implement DBSCAN clustering on the preprocessed data and visualize the resulting clusters.

```python
# Initialize and fit the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(data_scaled)

# Add the cluster labels to the original data
data['DBSCAN_Cluster'] = dbscan.labels_

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_scaled['Income'], y=data_scaled['MntWines'], hue=data['DBSCAN_Cluster'], palette='viridis')
plt.title('DBSCAN Clustering Results')
plt.xlabel('Income (Normalized)')
plt.ylabel('MntWines (Normalized)')
plt.show()
```

### **3.5 Model Performance and Evaluation**

In this section, we will compare the performance of the different clustering models (K-Means, Hierarchical Clustering, DBSCAN, and Clustering with Predictive Analytics) using appropriate evaluation metrics. The goal is to select the best model for further hyperparameter tuning.

**Evaluation Metrics for Clustering**

- **Silhouette Score**: Measures how similar an object is to its own cluster compared to other clusters. The score ranges from -1 to 1, where a higher value indicates better-defined clusters.

- **Davies-Bouldin Index**: Measures the average similarity ratio of each cluster with the cluster that is most similar to it. A lower Davies-Bouldin index indicates better clustering performance.

- **Calinski-Harabasz Index (Variance Ratio Criterion)**: Measures the ratio of the sum of between-cluster dispersion to the sum of within-cluster dispersion. A higher score indicates better-defined clusters.

- **Adjusted Rand Index (ARI)**: Measures the similarity between two data clusterings, considering all pairs of samples. The ARI is a measure of the agreement between two partitions, ignoring permutations and with chance normalization. A higher score indicates better clustering performance.

```python
def evaluate_clustering(labels, data_scaled):
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Count excluding noise if present
    if n_clusters > 1:  # Ensure there is more than one cluster
        sil_score = silhouette_score(data_scaled, labels)
        db_score = davies_bouldin_score(data_scaled, labels)
        ch_score = calinski_harabasz_score(data_scaled, labels)
    else:
        sil_score = -1  # Invalid score when only one cluster
        db_score = -1   # Invalid score when only one cluster
        ch_score = -1   # Invalid score when only one cluster
    return sil_score, db_score, ch_score
```

```python
# K-Means Clustering Evaluation
kmeans_labels = data['Cluster']
kmeans_sil, kmeans_db, kmeans_ch = evaluate_clustering(kmeans_labels, data_scaled)
```

```python
# Hierarchical Clustering Evaluation
from scipy.cluster.hierarchy import fcluster

# Use a predefined threshold to form flat clusters from the hierarchical clustering dendrogram
h_labels = fcluster(Z, t=5, criterion='maxclust')
h_sil, h_db, h_ch = evaluate_clustering(h_labels, data_scaled)
```

```python
# DBSCAN Clustering Evaluation
# Adjusting DBSCAN parameters for meaningful clustering
dbscan = DBSCAN(eps=0.3, min_samples=10)
dbscan.fit(data_scaled)
dbscan_labels = dbscan.labels_
dbscan_sil, dbscan_db, dbscan_ch = evaluate_clustering(dbscan_labels, data_scaled)
```

```python
# Clustering with Predictive Analytics Evaluation
logreg_labels = kmeans.predict(data_scaled)  # Using K-Means cluster labels as features in logistic regression
logreg_sil, logreg_db, logreg_ch = evaluate_clustering(logreg_labels, data_scaled)
```

### **3.7 Hyperparameter Tuning**

For K-Means Clustering, the primary hyperparameter that needs to be optimized is the number of clusters, K. Selecting the optimal number of clusters is essential to ensure that the clustering model accurately captures the underlying structure of the data, leading to well-defined and meaningful customer segments.

**Approach for Hyperparameter Tuning**

We will evaluate the performance of the K-Means Clustering model across a range of values for K, typically from 2 to 10. For each value of K, we will compute the Silhouette Score, Davies-Bouldin Index, and Calinski-Harabasz Index to assess the quality of the clusters. We will use these metrics to identify the optimal number of clusters. Additionally, we will use the Elbow Method, which involves plotting the within-cluster sum of squares (WCSS) against the number of clusters, to visually determine the optimal K.

```python
# Range of K values to evaluate
k_values = range(2, 11)

# Lists to store evaluation metrics
silhouette_scores = []
davies_bouldin_indices = []
calinski_harabasz_indices = []
wcss = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    labels = kmeans.labels_
    
    # Append WCSS for the Elbow Method
    wcss.append(kmeans.inertia_)
    
    # Compute evaluation metrics
    silhouette_scores.append(silhouette_score(data_scaled, labels))
    davies_bouldin_indices.append(davies_bouldin_score(data_scaled, labels))
    calinski_harabasz_indices.append(calinski_harabasz_score(data_scaled, labels))

# Plot the evaluation metrics
plt.figure(figsize=(15, 5))

# Silhouette Score
plt.subplot(1, 3, 1)
plt.plot(k_values, silhouette_scores, marker='o')
plt.title('Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')

# Davies-Bouldin Index
plt.subplot(1, 3, 2)
plt.plot(k_values, davies_bouldin_indices, marker='o')
plt.title('Davies-Bouldin Index vs. Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Davies-Bouldin Index')

# Calinski-Harabasz Index
plt.subplot(1, 3, 3)
plt.plot(k_values, calinski_harabasz_indices, marker='o')
plt.title('Calinski-Harabasz Index vs. Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Calinski-Harabasz Index')

plt.tight_layout()
plt.show()

# Plot the Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

Based on the above to determine the optimal number of clusters (K):

1. **Silhouette Score**:

- The Silhouette Score decreases as the number of clusters increases. The highest score is at K=2, and it gradually decreases.
- The optimal K should be where the Silhouette Score is relatively high.

2. **Davies-Bouldin Index**:

- The Davies-Bouldin Index is lowest at K=3 and increases with more clusters, indicating that the clusters are becoming less distinct and compact as K increases.
- The optimal K should be where the Davies-Bouldin Index is relatively low.

3. **Calinski-Harabasz Index**:

- The Calinski-Harabasz Index increases with the number of clusters, indicating that the dispersion between clusters is increasing.
- A higher score is better, but it should be balanced with the other metrics.

Considering the Silhouette Score, Davies-Bouldin Index, and Calinski-Harabasz Index together, K=4 appears to be a good choice for the number of clusters as it balances the metrics well in my view.
